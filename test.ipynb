```python
# Function to compare columns before merging and drop duplicates from the right dataframe
def merge_dataframes(df_left, df_right, key, suffixes=('_x', '_y')):
    overlapping_columns = [col for col in df_right.columns if col in df_left.columns and col != key]
    for col in overlapping_columns:
        if df_left[col].equals(df_right[col]):
            df_right = df_right.drop(col, axis=1)
    return df_left.merge(df_right, on=key, how='left', suffixes=suffixes)

# Using the function to merge df2 and df4
df2 = merge_dataframes(df2, df4, 'profile')
```
